```python
campos_obrigatorios_12 = [
    'FEBRE', 'MIALGIA', 'CEFALEIA', 'EXANTEMA',
    'VOMITO', 'NAUSEA', 'DOR_COSTAS', 'CONJUNTVIT', 'ARTRITE', 'ARTRALGIA', 'PETEQUIA_N',
    'LEUCOPENIA', 'LACO', 'DOR_RETRO', 'DIABETES', 'HEMATOLOG', 'HEPATOPAT', 'RENAL',
    'HIPERTENSA', 'ACIDO_PEPT', 'AUTO_IMUNE'
]
```

```python
def process_campos_obrigatorios(df, columns):
    for column in columns:
        if column in df.columns:
            df = df.dropna(subset=[column])
            df = df[df[column].isin([1, 2])]
    return df
```

```python
df = df[df['FEBRE'].isin([1, 2])]
```

```python
df = process_campos_obrigatorios(df, campos_obrigatorios_12)
```

```python
df = df[pd.to_numeric(df['CLASSI_FIN'], errors='coerce').notna() | df['CLASSI_FIN'].isna()]
```

```python
# SOROTIPO: Obrigatório Se campo 'RESUL_VI_N' ou campo 'RESUL_PCR_'= 1.
# Campo desabilitado se campo 'RESUL_VI_N' e campo 'RESUL_PCR_' diferente de 1.
# Habilitado apenas se campo 'ID_AGRAVO' = 1

df_sorotipo =  df[~(((df['RESUL_VI_N'] == 1) | (df['RESUL_PCR_'] == 1)) & df['SOROTIPO'].isna())]
df_sorotipo = df_sorotipo[~((df_sorotipo['SOROTIPO'].notna()) & (df_sorotipo['RESUL_VI_N'] != 1) & (df_sorotipo['RESUL_PCR_'] != 1))]
```

```python
# CLASSI_FIN: Campo Obrigatório quando o campo 'DT_ENCERRA' estiver preenchido.
# Desabilitar categoria 13 se campo 'ID_AGRAVO'= 1.
# Desabilitar categorias 10, 11 e 12 se campo 'ID_AGRAVO'=2

df_classifin = df_sorotipo[~(df_sorotipo['DT_ENCERRA'].notna() & df_sorotipo['CLASSI_FIN'].isna())]
df_classifin = df_classifin[~((df_classifin['ID_AGRAVO'] == 1) & (df_classifin['CLASSI_FIN'] == 13))]
df_classifin = df_classifin[~((df_classifin['ID_AGRAVO'] == 2) & (df_classifin['CLASSI_FIN'].isin([10, 11, 12])))]
```

```python
# CLINIC_CHIK: Campo Obrigatório
# Se campo 'CLASSI_FIN'=13 e se campo 'CRITERIO'= 1 ou 2

df_clinicchik = df_classifin[~((df_classifin['CLASSI_FIN'] == 13) & (df_classifin['CRITERIO'].isin([1, 2])) & df_classifin['CLINC_CHIK'].isna())]
```

```python
# DT_OBITO: Campo Obrigatório se 'EVOLUCAO'=2 ou 3 ou 4.
# Data deve ser maior ou igual a 'DT_ALRM'.

df_dtobito= df_clinicchik[~(((df_clinicchik['EVOLUCAO'].isin([2, 3, 4])) & (df_clinicchik['DT_OBITO'].isna() | (df_clinicchik['DT_OBITO'] < df_clinicchik['DT_ALRM']))))]
df_dtobito
```

```python
#DT_ENCERRA: Campo Obrigatório quando campo 'CLASSI_FIN' estiver preenchido E o campo 'CRITERIO' for diferente de 3.
# Campo desabilitado se o campo 'CRITERIO' for igual a 3.

df_dtobito['DT_ENCERRA'] = pd.to_datetime(df_dtobito['DT_ENCERRA'], errors='coerce')
df_encerra = df_dtobito[~((df_dtobito['CLASSI_FIN'].notna()) & (df_dtobito['CRITERIO'] != 3) & df_dtobito['DT_ENCERRA'].isna())]
df_encerra = df_encerra[~((df_encerra['CRITERIO'] == 3) & df_encerra['DT_ENCERRA'].notna())]

df_encerra
```

```python
def alrm_condicional(df):
    # Condições para os campos obrigatórios
    mandatory_conditions = (df['CLASSI_FIN'].isin([11, 12]) & df['CRITERIO'].isin([1, 2]))

    # Lista de campos a serem verificados
    fields = ['ALRM_SANG', 'ALRM_HEMAT', 'ALRM_HEPAT', 'ALRM_LIQ', 'ALRM_LETAR', 'ALRM_ABDOM', 'ALRM_VOM', 'ALRM_PLAQ', 'ALRM_HIPOT', 'DT_ALRM']

    for field in fields:
        # Remover linhas onde o campo é obrigatório mas está nulo
        df = df[~(mandatory_conditions & df[field].isna())]
        # Remover linhas onde o campo está preenchido mas a condição não é satisfeita
        df = df[~((~mandatory_conditions) & df[field].notna())]

    return df

```

```python
df_almr = alrm_condicional(df_encerra)
df_almr
```

```python
def grav_condicional(df):
    # Condições para os campos obrigatórios
    mandatory_conditions = (df['CLASSI_FIN'] == 12) & (df['CRITERIO'].isin([1, 2]))

    # Lista de campos a serem verificados
    fields = [
        'GRAV_PULSO', 'GRAV_CONV', 'GRAV_ENCH', 'GRAV_INSUF', 'GRAV_TAQUI', 
        'GRAV_EXTRE', 'GRAV_HIPOT', 'GRAV_HEMAT', 'GRAV_MELEN', 'GRAV_METRO', 
        'GRAV_SANG', 'GRAV_AST', 'GRAV_MIOC', 'GRAV_CONSC', 'GRAV_ORGAO', 'DT_GRAV'
    ]

    for field in fields:
        # Remover linhas onde o campo é obrigatório mas está nulo
        df = df[~(mandatory_conditions & df[field].isna())]
        # Remover linhas onde o campo está preenchido mas a condição não é satisfeita
        df = df[~((~mandatory_conditions) & df[field].notna())]

    return df
```

```python
# 35 - Exame sorológico (IgM) Chikungunya Soro 1 Data da Coleta 
# DT_CHIK_S1: Campo Essencial
# DT_CHIK_S1 >= DT_ALRM
# Habilitado apenas se campo ‘ID_AGRAVO’ = 2

df_grav['DT_CHIK_S1'] = pd.to_datetime(df_grav['DT_CHIK_S1'], errors='coerce')
df_grav['DT_ALRM'] = pd.to_datetime(df_grav['DT_ALRM'], errors='coerce')

# Condições para os campos obrigatórios
chik_condition = (df_grav['ID_AGRAVO'] == 2)

# Remover linhas onde DT_CHIK_S1 é nulo ou menor que DT_ALRM quando ID_AGRAVO é 2

# Remover linhas onde DT_CHIK_S1 é nulo ou menor que DT_ALRM quando ID_AGRAVO é 2
df_grav = df_grav[~((chik_condition & df_grav['DT_CHIK_S1'].isna()) | (chik_condition & (df_grav['DT_CHIK_S1'] < df_grav['DT_ALRM'])))]

# Remover linhas onde DT_CHIK_S1 está preenchida mas ID_AGRAVO é diferente de 2
df_chiks1 = df_grav[~((~chik_condition) & df_grav['DT_CHIK_S1'].notna())]
```

```python
# 36 - Exame sorológico (IgM) Chikungunya Soro 2 Data da Coleta 
# DT_CHIK_S2: Campo Essencial 
# DT_CHIK_S2 >= DT_CHIK_S1 
# Habilitado apenas se campo ‘ID_AGRAVO’ = 2


df_chiks1['DT_CHIK_S2'] = pd.to_datetime(df_chiks1['DT_CHIK_S2'], errors='coerce')

# Condições para os campos obrigatórios
chik_condition = (df_chiks1['ID_AGRAVO'] == 2)

# Remover linhas onde DT_CHIK_S2 é nulo ou menor que DT_CHIK_S1 quando ID_AGRAVO é 2
df_chiks1 = df_chiks1[~((chik_condition & df_chiks1['DT_CHIK_S2'].isna()) | (chik_condition & (df_chiks1['DT_CHIK_S2'] < df_chiks1['DT_CHIK_S1'])))]

# Remover linhas onde DT_CHIK_S2 está preenchida mas ID_AGRAVO é diferente de 2
df_chiks2 = df_chiks1[~((~chik_condition) & df_chiks1['DT_CHIK_S2'].notna())]
```

```python
def data_unabled(df):
    # Converter colunas de data para o formato datetime
    df['DT_PRNT'] = pd.to_datetime(df['DT_PRNT'], errors='coerce')

    # Condições para os campos obrigatórios
    chik_condition = (df['ID_AGRAVO'] == 2)

    # Lista de campos a serem verificados
    fields_to_check = ['DT_PRNT', 'RES_CHIKS1', 'RES_CHIKS2', 'RESUL_PRNT']

    for field in fields_to_check:
        # Remover linhas onde o campo é obrigatório mas está nulo
        df = df[~(chik_condition & df[field].isna())]
        # Remover linhas onde o campo está preenchido mas a condição não é satisfeita
        df = df[~((~chik_condition) & df[field].notna())]

    return df

df_unabled = data_unabled(df_chiks2)
```

```python
def idagravo_abled(df):

    # Condições para os campos obrigatórios
    chik_condition = (df['ID_AGRAVO'] == 2)
    dengue_condition = (df['ID_AGRAVO'] == 1)

    # Lista de campos a serem verificados e suas condições
    fields_conditions = [
        ('DT_PRNT', chik_condition),
        ('RES_CHIKS1', chik_condition),
        ('RES_CHIKS2', chik_condition),
        ('RESUL_PRNT', chik_condition),
        ('DT_SORO', dengue_condition & (df['DT_SORO'] >= df['DT_ALRM'])),
        ('RESUL_SORO', dengue_condition),
        ('DT_NS1', dengue_condition & (df['DT_NS1'] >= df['DT_ALRM'])),
        ('RESUL_NS1', dengue_condition),
        ('HISTOPA_N', dengue_condition),
        ('IMUNOH_N', dengue_condition)
    ]

    for field, condition in fields_conditions:
        # Remover linhas onde o campo é obrigatório mas está nulo
        df = df[~(condition & df[field].isna())]
        # Remover linhas onde o campo está preenchido mas a condição não é satisfeita
        df = df[~((~condition) & df[field].notna())]

    return df

df_agravo = idagravo_abled(df_unabled)
```

```python
def data_pcr(df):
    # Converter colunas de data para o formato datetime
    df['DT_PRNT'] = pd.to_datetime(df['DT_PRNT'], errors='coerce')
    df['DT_SORO'] = pd.to_datetime(df['DT_SORO'], errors='coerce')
    df['DT_NS1'] = pd.to_datetime(df['DT_NS1'], errors='coerce')
    df['DT_ALRM'] = pd.to_datetime(df['DT_ALRM'], errors='coerce')
    df['DT_VIRAL'] = pd.to_datetime(df['DT_VIRAL'], errors='coerce')
    df['DT_PCR'] = pd.to_datetime(df['DT_PCR'], errors='coerce')

    # Condições para os campos obrigatórios
    chik_condition = (df['ID_AGRAVO'] == 2)
    dengue_condition = (df['ID_AGRAVO'] == 1)
    viral_pcr_condition = (df['ID_AGRAVO'].isin([1, 2]))

    # Lista de campos a serem verificados e suas condições
    fields_conditions = [
        ('DT_PRNT', chik_condition),
        ('RES_CHIKS1', chik_condition),
        ('RES_CHIKS2', chik_condition),
        ('RESUL_PRNT', chik_condition),
        ('DT_SORO', dengue_condition & (df['DT_SORO'] >= df['DT_ALRM'])),
        ('RESUL_SORO', dengue_condition),
        ('DT_NS1', dengue_condition & (df['DT_NS1'] >= df['DT_ALRM'])),
        ('RESUL_NS1', dengue_condition),
        ('HISTOPA_N', dengue_condition),
        ('IMUNOH_N', dengue_condition),
        ('DT_VIRAL', viral_pcr_condition & (df['DT_VIRAL'] >= df['DT_ALRM'])),
        ('RESUL_VI_N', viral_pcr_condition),
        ('DT_PCR', viral_pcr_condition & (df['DT_PCR'] >= df['DT_ALRM'])),
        ('RESUL_PCR_', viral_pcr_condition)
    ]

    for field, condition in fields_conditions:
        # Remover linhas onde o campo é obrigatório mas está nulo
        df = df[~(condition & df[field].isna())]
        # Remover linhas onde o campo está preenchido mas a condição não é satisfeita
        df = df[~((~condition) & df[field].notna())]

    return df

# Aplicar a função de validação e limpeza
df_pcr = data_pcr(df_agravo)
```
